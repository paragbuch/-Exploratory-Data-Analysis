```python
total_player = df5.groupby(['Position'], as_index=False).count()[['Name','Position']]
total_player.columns=['Total Players','Position']


total_player
```

```python
fig, axx =plt.subplots(figsize=(12,6))
axx.set_title("Total Players by their Position in PKL")
sns.barplot(x='Total Players', y='Position',hue='Position', data=total_player, palette='Set1')
```

```python
pos=['All Rounder','Raider','Defender']
pos_play = df5.loc[df5['Position'].isin(pos)]
pos_play

# Find total number of Position

total_pos = pos_play.groupby(['Position'], as_index=False).count()[['Position','Name']]
total_pos.columns=['Position','Total Players']

# Find total position count of each type

top5pos = pos_play.groupby(['Position','Name'], as_index=False)[['Position']].count()
top5pos.columns=['Name', 'Position']

top5pos
```
